```python
import pandas as pd
```

```python
file_path = '/Users/noahdimonti/Documents/Uni/MPhil in Eng and CS/EV Apartment Building Project/Modelling Data/Potential Datasets/ev-energy-consumption-and-moves-ricardo-et-al/Dataset.xlsx'
load = pd.read_excel(file_path, sheet_name='Load')
```

```python
load['total_load'] = load.sum(axis=1, numeric_only=True)
load
```

```python
load_profile = pd.DataFrame(pd.date_range(start='2019-01-01 00:00:00', periods=96, freq='15min'), columns=['timestamp'])
load_profile['load'] = load['total_load']
load_profile
```

```python
load_profile_1 = load_profile
load_profile_2 = load_profile
load_profile_1
```

```python
load_profile_2.timestamp = pd.date_range(start='2019-01-02 00:00:00', periods=96, freq='15min')
merged = pd.concat([load_profile_1, load_profile_2])
merged
```

```python
timeseries = pd.DataFrame(pd.date_range(start='2019-01-01 00:00:00', periods=96*2, freq='15min'), columns=['timestamp'])
timeseries['load'] = load_profile.load
timeseries
```

```python
timeseries['load'].iloc[96:] = load_profile.load
timeseries
```

```python
timeseries.set_index('timestamp', inplace=True)
timeseries.to_csv('/Users/noahdimonti/Developer/PycharmProjects/EV_model_test/load_profile_2_days.csv')
```

```python
merged.set_index('timestamp', inplace=True)
merged.to_csv('/Users/noahdimonti/Developer/PycharmProjects/EV_model_test/load_profile_2_days.csv')
```
